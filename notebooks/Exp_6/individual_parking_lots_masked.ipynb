```python
from planet_img_utils import *
from model_train_utils import *
import os
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
from tqdm.auto import tqdm
import warnings
warnings.filterwarnings('ignore')
```

```python
path2data = '/home/akansh-i2sc/Desktop/Project/detect-car-LR/data/data_parking_masked/'
os.listdir(path2data)

paths = []
for i in os.listdir(path2data):
    for j in os.listdir(path2data+i):
        img_path = path2data+i+'/'+j+'/**/*.tif'
        paths.append(img_path)
```

```python
selceted_paths_masked = []
selceted_paths_square = []
```

```python
for path in tqdm(paths):
    clean_1 = clean_data_by_size(path)
    clean_2 = clean_data_by_black_pixels(clean_1)
    clean_3 = clean_data_by_white_pixels(clean_2)
    masked_clean = [i for i in clean_3 if 'Masked' in i]
    square_clean = [i for i in clean_3 if 'Planet' in i]

    if len(masked_clean) != 0:
        df_masked = label_csv(masked_clean)
        model_mask, train_loss_msk, train_accuracy_msk, train_auc_msk, test_loss_msk, test_accuracy_msk, test_auc_msk = train(df_masked, train_transform, test_transform, test_size = 0.2, num_epochs = 100, save_path= None, batch_size = 64)
        if np.sum(np.array(test_auc_msk) > 0.5) > 5: 
            selceted_paths_masked.append(path)

    if len(square_clean) != 0:
        df_square = label_csv(square_clean)
        model_square, train_loss_sq, train_accuracy_sq, train_auc_sq, test_loss_sq, test_accuracy_sq, test_auc_sq = train(df_square, train_transform, test_transform, test_size = 0.2, num_epochs = 100, save_path= None, batch_size = 64)
        if np.sum(np.array(test_auc_sq) > 0.5) > 5:
            selceted_paths_square.append(path)
```
